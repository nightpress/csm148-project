## Question 1 - explain dataset chosen

The dataset our team has chosen for the project is the spotify dataset. We chose the spotify data because the dataset appears to be the easiest to work with and music is a shared interest with everyone in the group.

```python
# this cell may take up to 30 seconds to run to convert the Hugginface Dataset class to Pandas DataFrame for better EDA

from datasets import load_dataset
import pandas as pd

HF_SPOTIFY_DATASET_PATH = "maharshipandya/spotify-tracks-dataset"
# https://huggingface.co/datasets/maharshipandya/spotify-tracks-dataset

ds = load_dataset(HF_SPOTIFY_DATASET_PATH) #only has 'train' split as key
df = pd.DataFrame(ds["train"])
df.head()
```

```python
# get dataset info
df.info()
df.describe()
df.head()
```

```python
print(df.columns)
```

```python
#
cols = ['Unnamed: 0', 'track_id', 'artists', 'album_name', 'track_name',
       'popularity', 'duration_ms', 'explicit', 'danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature',
       'track_genre']

for column in cols:
  print("Unique values of", column, ":", df[column].unique())
```

```python
# find missing values. only one example missing track_id, artist and album_name
df.isnull().sum()
df[df.isnull().any(axis=1)]
```

```python
df = df.dropna()
```

#Question 4
One EDA that we applied to our data is a heatmap. This gives us the best idea of the correlation with our target variable, as well as the correlation between our feature variables. Finding the correlation is a crucial way to dterermine the best predictors.

```python
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10,10))
for i, col in enumerate(corr):
    cols = list(corr.columns)
    cols.remove(col)
    axes[i//4, i%4].plot(corr[col].drop(col), cols)
    axes[i//4, i%4].set_title(f"Corr with {col}")


plt.tight_layout()
plt.show()
```
